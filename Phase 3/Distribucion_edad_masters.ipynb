```python
import pandas as pd
from scipy import stats
import numpy as np 
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("train.csv")

df
```

```python
patron = r'Master'
niños = df[df['Name'].str.contains(patron)]

niños
```

```python
edades = niños["Age"].to_numpy()
edad_max = max(edades)
edad_min = min(edades)
limpia_edades = edades[~np.isnan(edades)]
limpia_edades
```

Prueba de normalidad Anderson-Darling

```python
stats.anderson(limpia_edades, dist="norm")

```

No sigue una distribución normal, asi que haremos una prueba para la distribución exponencial 

```python
stats.anderson(limpia_edades, dist="expon")
```

Aparantemente si sigue una distribución exponencial 

```python
plt.hist(edades,bins=np.arange(edad_min,edad_max))
```
