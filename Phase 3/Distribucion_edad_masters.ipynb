```python
import pandas as pd
from scipy import stats
import numpy as np 
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("train.csv")

df
```

```python
patron = r'Master'
niÃ±os = df[df['Name'].str.contains(patron)]

niÃ±os
```

```python
edades = niÃ±os["Age"].to_numpy()
edad_max = max(edades)
edad_min = min(edades)
limpia_edades = edades[~np.isnan(edades)]
limpia_edades
```

Prueba de normalidad Anderson-Darling

```python
stats.anderson(limpia_edades, dist="norm")

```

No sigue una distribuciÃ³n normal, asi que haremos una prueba para la distribuciÃ³n exponencial 

```python
stats.anderson(limpia_edades, dist="expon")
```

Aparantemente si sigue una distribuciÃ³n exponencial 

```python
plt.hist(edades,bins=np.arange(edad_min,edad_max))
```
