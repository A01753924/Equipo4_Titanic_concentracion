### Load Data

```python
import pandas as pd
```

```python
customers = pd.read_csv("customers_sampled.csv")
customers.head()
```

```python
products = pd.read_csv("20230223_productos.csv")
products.rename(columns={'Material': 'material'}, inplace=True)
products.head()
```

```python
sales = pd.read_csv("ventas.csv")
sales.head()
```

### Select Information

```python
columns = ["material", "Brand", "Returnability", "Flavor", "Container", "ProductCategory"]
products = products[columns]
products.head()
```

```python
columns = ["CustomerId", "sub_canal_comercial", "parques", "supermercados", "hospitales", "preescolares", "primarias", "secundarias", "preparatorias", "universidades", "gimnasios"]
customers = customers[columns]
customers.head()
```

```python
sales.drop("uni_box", axis = 1, inplace = True)
sales.head()
```

### Link Data

```python
merged_data = pd.merge(sales, customers, on='CustomerId', how='left')
merged_data = pd.merge(merged_data, products, on='material', how='left')
merged_data.head()
```

```python
merged_data.shape
```

```python
merged_data[(merged_data["CustomerId"] == 499920078) & (merged_data["material"] == 14050)]["calmonth"].count()
```

```python
counts = merged_data.groupby(['CustomerId', 'material'])['calmonth'].transform('count')
merged_data['count_greater_than_5'] = counts >= 5
unique_combinations = merged_data.drop_duplicates(subset=['CustomerId', 'material'])
```

```python
unique_combinations.drop("calmonth", axis = 1, inplace = True)
unique_combinations.head()
```

```python
unique_combinations.shape
```

### Manage Categorical Data

```python
from sklearn.preprocessing import OneHotEncoder, LabelBinarizer
```

```python
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded_data = encoder.fit_transform(unique_combinations[['sub_canal_comercial', 'Brand', 'Flavor', 'Container', 'ProductCategory']])
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['sub_canal_comercial', 'Brand', 'Flavor', 'Container', 'ProductCategory']))
semi_final_df = pd.concat([unique_combinations, encoded_df], axis=1)
semi_final_df.drop(['sub_canal_comercial', 'Brand', 'Flavor', 'Container', 'ProductCategory'], axis = 1, inplace = True)
semi_final_df.head()
```

```python
semi_final_df.dropna(inplace = True)
```

```python
lb = LabelBinarizer()
semi_final_df['Returnability_binary'] = lb.fit_transform(semi_final_df['Returnability'])
semi_final_df.drop(["Returnability", "material"], axis = 1, inplace = True)
semi_final_df.head()
```

### Data Semgentaion

```python
from sklearn.model_selection import train_test_split
```

```python
X = semi_final_df.drop("count_greater_than_5", axis = 1)
y = semi_final_df["count_greater_than_5"].astype(int)
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

### Train Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(random_state=0).fit(X_train, y_train)
```

### Evaluate Model

```python
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```
