### Load Data

```python
import pandas as pd
```

```python
products = pd.read_csv("20230223_productos.csv")
products.rename(columns={'Material': 'material'}, inplace=True)
products.head()
```

```python
sales = pd.read_csv("ventas.csv")
sales.head()
```

### Select Information

```python
columns = ["material", "Brand", "Returnability", "Flavor", "Container", "ProductCategory"]
products = products[columns]
products.head()
```

```python
sales.drop("uni_box", axis = 1, inplace = True)
sales.head()
```

### Link Data

```python
merged_data = pd.merge(sales, products, on='material', how='left')
merged_data.head()
```

```python
counts = merged_data.groupby(['CustomerId', 'material'])['calmonth'].transform('count')
merged_data['count_greater_than_5'] = counts >= 5
unique_combinations = merged_data.drop_duplicates(subset=['CustomerId', 'material'])
```

```python
unique_combinations.drop("calmonth", axis = 1, inplace = True)
unique_combinations.head()
```

```python
unique_combinations.shape
```

### Manage Categorical Data

```python
encoded_data = pd.get_dummies(unique_combinations, columns = ['Returnability', 'Flavor', 'Container', 'Brand', 'ProductCategory'])
encoded_data.head()
```

```python
encoded_data.dropna(inplace = True)
```

### Data Semgentaion

```python
from sklearn.model_selection import train_test_split
```

```python
clusters = pd.read_csv('Clustered_Customers.csv')
clusters.head()
```

```python
grouped_data = pd.merge(encoded_data, clusters, on='CustomerId', how='left')
grouped_data.head()
```

```python
group0 = grouped_data[grouped_data['Cluster'] == 3]
group0.info()
```

```python
group0.drop(['CustomerId', 'material'], axis = 1, inplace = True)
X = group0.drop("count_greater_than_5", axis = 1)
y = group0["count_greater_than_5"].astype(int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

### Train Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(random_state=0).fit(X_train, y_train)
```

### Evaluate Model

```python
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
y_pred = rf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression(random_state=0).fit(X, y)
```

```python
y_pred = logreg.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn import svm
svm_clf = svm.LinearSVC(max_iter = 5000).fit(X, y)
```

```python
y_pred = svm_clf.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
import tensorflow as tf
```

```python
X_train = tf.constant(X_train)
y_train = tf.constant(y_train)
X_test = tf.constant(X_test)
y_test = tf.constant(y_test)
```

```python
# Create the model
model = tf.keras.Sequential([
                  tf.keras.layers.Dense(128, activation="relu"),
                  tf.keras.layers.Dense(256, activation="relu"),
                  tf.keras.layers.Dense(128, activation="relu"),
                  tf.keras.layers.Dense(1, activation="sigmoid")
 ])
# Compile the model
model.compile(loss="binary_crossentropy",
              optimizer="Adam",
              metrics=["accuracy"])
# Fit the model
history = model.fit(X_train,
                    y_train,
                    epochs=50,
                    validation_data=(X_test, y_test),
                    verbose=0)
model.evaluate(X_test, y_test)
```
